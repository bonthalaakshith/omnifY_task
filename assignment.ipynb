# Return on investment

```python
df['Week'] = pd.to_datetime(df['Week'])
```

```python
df['Month'] = df['Week'].dt.month
```

```python
df['Month'] = df['Week'].dt.strftime('%B')
```

```python
df = df.sort_values(by='ROI', ascending=False)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(df['Month'], df['ROI'], color='skyblue')
plt.xlabel('Month')
plt.ylabel('Return on Investment (ROI)')
plt.title('ROI by Month')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
profit_by_keyword=df.groupby('Search Keyword')['ROI'].sum()
profit_by_keyword=profit_by_keyword.idxmax()
profit_by_keyword
```

```python
weekly_report = df.groupby('Week').agg(
    Spends_weekly=('Cost ($)', 'sum'),
    Returns_weekly=('cost_per_click', 'sum'),
    Returns_percentage_weekly=('cost_per_click', lambda x: x.sum() / df['Cost ($)'].sum() * 100)
).reset_index()

monthly_report = df.groupby('Month').agg(
    Spends_monthly=('Cost ($)', 'sum'),
    Returns_monthly=('cost_per_click', 'sum'),
    Returns_percentage_monthly=('cost_per_click', lambda x: x.sum() / df['Cost ($)'].sum() * 100)
).reset_index()
```

```python
print("Weekly Report:")
print(weekly_report)

print("\nMonthly Report:")
print(monthly_report)
```

# Create a combined report with all key metrics (esp: Spends and Returns %) in a weekly and monthly format. Decide the number of data fields you would like to showcase here.

```python
df1["Date of Report"]=pd.to_datetime(df1["Date of Report"])
df1.head()
```

```python
df1['Week'] = df1['Date of Report'].dt.to_period('W').apply(lambda r: r.start_time)
df1['Month'] = df1['Date of Report'].dt.to_period('M').apply(lambda r: r.start_time)
```

```python
weekly_report = df1.groupby('Week').agg(
    Spends_weekly=('Money Spent ($)', 'sum'),
    Returns_weekly=('ClicksPerDollar', 'sum'),
    Returns_percentage_weekly=('ClicksPerDollar', lambda x: x.sum() / df1['Money Spent ($)'].sum() * 100)
).reset_index()

monthly_report = df1.groupby('Month').agg(
    Spends_monthly=('Money Spent ($)', 'sum'),
    Returns_monthly=('ClicksPerDollar', 'sum'),
    Returns_percentage_monthly=('ClicksPerDollar', lambda x: x.sum() / df1['Money Spent ($)'].sum() * 100)
).reset_index()
```

```python
print("Weekly Report:")
print(weekly_report)

print("\nMonthly Report:")
print(monthly_report)
```
